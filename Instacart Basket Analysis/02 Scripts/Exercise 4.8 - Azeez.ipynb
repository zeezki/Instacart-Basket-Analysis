#### 1. Creating loyalty flag

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts()
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
ords_prods_merge.groupby('user_id')['prices'].mean()
```

#### 2. Creating spenders type

```python
ords_prods_merge['Average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge
```

```python
ords_prods_merge.loc[ords_prods_merge['Average_price'] < 10, 'spenders_type'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['Average_price'] >= 10, 'spenders_type'] = 'High spender'
```

```python
ords_prods_merge.head()
```

#### 3. Creating order frequency flag

```python
ords_prods_merge['order_frequency_flag'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency_flag'] > 20, 'customers_frequecy'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency_flag'] > 10) & (ords_prods_merge['order_frequency_flag'] <= 20), 'customers_frequecy'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency_flag'] <= 10, 'customers_frequecy'] = 'Frequent customer'
```

```python
ords_prods_merge['customers_frequecy'].value_counts()
```
