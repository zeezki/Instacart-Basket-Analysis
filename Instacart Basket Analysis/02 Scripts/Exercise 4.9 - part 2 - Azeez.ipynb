#### 1. Bar chart

```python
bar= customers_new['orders_day_of_week'].value_counts().plot.bar()
plt.xlabel('Day of the week')
plt.ylabel('Total orders')
plt.title('Days with most orders')
bar.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x) ))
plt.xticks([0,1,2,3,4,5,6],['Saturday', 'Sunday', 'Monday', 'Friday', 'Thursday', 'Tuesday', 'Wednesday'], rotation=45)
plt.figure(figsize=(20,10))
plt.savefig("output.png", bbox_inches="tight")
plt.show()
```
